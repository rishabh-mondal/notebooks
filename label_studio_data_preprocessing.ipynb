```python
import os
import os
import shutil
```

```python
# Define the directory containing the .txt files
source_directory = "../data/gms_data/delhi_ncr_data/train_obb_data/labels/"
destination_directory = "../data/gms_data/delhi_ncr_data/train_obb_data/labels_1/"
```

```python



# Ensure the destination directory exists
os.makedirs(destination_directory, exist_ok=True)

# Dictionary to track renamed files
renamed_files = {}

# Rename the files and check for common filenames
common_files = []
for filename in os.listdir(source_directory):
    if filename.endswith(".txt"):
        new_filename = filename.split('-', 1)[-1]
        print(f"Renaming {filename} to {new_filename}")
        old_file_path = os.path.join(source_directory, filename)
        new_file_path = os.path.join(destination_directory, new_filename)
        
        # Check if the new filename already exists in the renamed_files dictionary
        if new_filename in renamed_files:
            common_files.append(new_filename)
            print(f"Common file found: {new_filename}")
        else:
            renamed_files[new_filename] = old_file_path
            shutil.copy2(old_file_path, new_file_path)  # Copy renamed file to the destination

# Print common files and save the list to a file
if common_files:
    print("Common files after renaming:")
    for file in common_files:
        print(file)
    with open(os.path.join(destination_directory, "common_files.txt"), "w") as f:
        for file in common_files:
            f.write(f"{file}\n")
else:
    print("No common files found after renaming.")

print("Renamed files have been saved to the destination directory.")
```

```python
import os
from collections import defaultdict

# Define the directory path
directory_path = '/home/rishabh.mondal/Brick-Kilns-project/ijcai_2025_kilns/data/gms_data/delhi_ncr_brick_kilns'

# Dictionary to store file names and their counts
file_name_count = defaultdict(int)

# Walk through the directory
for root, dirs, files in os.walk(directory_path):
    for file in files:
        if file.endswith('.tif'):
            file_name_count[file] += 1

# Check for common file names
common_files = [file for file, count in file_name_count.items() if count > 1]

# Print the result
if common_files:
    print(f"Found {len(common_files)} duplicate .tif file names:")
    for file in common_files:
        print(file)
else:
    print("No duplicate .tif file names found.")
```
