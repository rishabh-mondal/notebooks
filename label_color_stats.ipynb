```python
import pandas as pd
```

```python

bbox_df=f"../result_csv/state_rgb_values.csv"
df = pd.read_csv(bbox_df)
display(df)
```

```python
## sort by state using average red value
df = df.sort_values(by='Average Red', ascending=False)
display(df)
```

```python
background_df=f"../result_csv/state_background_rgb_values.csv"
df2 = pd.read_csv(background_df)
display(df2)
```

```python
df2 = df2.sort_values(by='Average Green', ascending=False)
display(df2.style.background_gradient(cmap='viridis', subset=['Average Green']))
```

```python
mAP = {
    'west_bengal': 0.460848,
    'uttar_pradesh': 0.377995,
    'haryana': 0.318754,
    'punjab': 0.406539,
    'test_bihar': 0.459792,
    'bihar':  None

}

df2['mAP'] = df2['State'].apply(lambda x: mAP[x])
display(df2)
df2.to_csv(f"../result_csv/state_background_rgb_values.csv", index=False)
```

```python
df['mAP'] = df['State'].apply(lambda x: mAP[x])
display(df)
df.to_csv(f"../result_csv/state_rgb_values.csv", index=False)
```

```python
display(df.style.background_gradient(cmap='viridis', subset=['Average Red','mAP']))
display(df2.style.background_gradient(cmap='viridis', subset=['Average Green','mAP']))
```

```python
## find out the correlation between the average red value and mAP
print(df['Average Red'].corr(df['mAP']))
print(df['Average Green'].corr(df['mAP']))
print(df['Average Blue'].corr(df['mAP']))
print(df2['Average Red'].corr(df2['mAP']))
print(df2['Average Green'].corr(df2['mAP']))
print(df2['Average Blue'].corr(df2['mAP']))
```
